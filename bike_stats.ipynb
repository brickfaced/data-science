# 3. Given all the weather data here, find the average precipitation per month, and the median precipitation.

```python
weather_df['Date'] = pd.to_datetime(weather_df['Date'])
weather_df['Month'] = weather_df['Date'].dt.month
pd.DataFrame(weather_df['Precipitation_In'].groupby(weather_df.Month).mean())
```

```python
pd.DataFrame(weather_df['Precipitation_In'].groupby(weather_df.Month).median())
```

# 5. When a bike station is modified, is it more likely that it’ll lose bikes or gain bikes? How do you know?

```python
diff = station_df.dropna(subset=['modification_date'])
```

```python
diff.count()
```

```python
diff[diff['install_dockcount'] > diff['current_dockcount']].count()
```

It is more likely because there is more instances where there is less bikes after modification date.
